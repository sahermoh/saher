```python
import numpy as np
import cv2
from matplotlib import pyplot as plt


#tramform the image to ttf>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>


img = cv2.imread('oso.jpg',0)
dft = cv2.dft(np.float32(img),flags = cv2.DFT_COMPLEX_OUTPUT)
dft_shift = np.fft.fftshift(dft)




magnitude_spectrum = 20*np.log(cv2.magnitude(dft_shift[:,:,0],dft_shift[:,:,1]))
os = cv2.imread('oso..jpg',0)
# create a mask first, band reject filter

rows, cols = img.shape# here we defind number of cals and rows in image 
crow,ccol = rows/2 , cols/2# define the center if rows and cols
r=160 #cutoff of low pass ideal filter
lp = np.zeros(img.size, img.dtype).reshape(img.shape)#create mask with size of the image
mask = np.zeros((rows,cols,2),np.uint8)
for icounter in range(1, rows):
    for jcounter in range(1, cols):
        if ((icounter - crow)**2 + (jcounter - ccol)**2) < r**2:
            mask[icounter, jcounter] = 1
            lp[icounter, jcounter] = 1
           
        
       
    
            
            

            


plt.imshow(lp, cmap = 'gray')#show the low pass filter
plt.title('Filter'), plt.xticks([]), plt.yticks([])
plt.show() 


r2=150#cutoff of high pass filter
hp = np.zeros(img.size, img.dtype).reshape(img.shape)#crete mask with size of image 
mask2 = np.zeros((rows,cols,2),np.uint8)
for icounter in range(1, rows):
    for jcounter in range(1, cols):
        if ((icounter - crow)**2 + (jcounter - ccol)**2) > r2**2:
            mask2[icounter, jcounter] = 1
            hp[icounter, jcounter] = 1
            
            
            
            

plt.imshow(hp, cmap = 'gray')#show the high pass filter
plt.title('Filter'), plt.xticks([]), plt.yticks([])
plt.show()    
reject = np.zeros(img.size, img.dtype).reshape(img.shape)
maskre = np.zeros((rows,cols,2),np.uint8)




reject=lp*hp#find band bass filter by muliple of low pass with high pass
maskre=mask*mask2
plt.imshow(reject, cmap = 'gray')#show the band pass filter
plt.title('Filter'), plt.xticks([]), plt.yticks([])
plt.show()




k=cv2.bitwise_not(reject)#not operation because we need to reject all the frequancy that inside of the circle not outside
kk=cv2.bitwise_not(maskre)

plt.imshow(k, cmap = 'gray')
plt.title('Filter'), plt.xticks([]), plt.yticks([])
plt.show() 






# apply mask and inverse DFT
fshift = dft_shift*kk
f_ishift = np.fft.ifftshift(fshift)
img_back = cv2.idft(f_ishift)
img_back = cv2.magnitude(img_back[:,:,0],img_back[:,:,1])






#plt.imshow(img, cmap = 'gray')
#plt.title('Input Image'), plt.xticks([]), plt.yticks([])
#plt.imshow(img, cmap = 'gray')
#plt.title(' input image'), plt.xticks([]), plt.yticks([])
#plt.show()
fig, axes = plt.subplots(ncols=3, figsize=(8, 2.5))
ax = axes.ravel()
ax[0] = plt.subplot(1, 3, 1)
ax[1] = plt.subplot(1, 3, 2)
ax[2] = plt.subplot(1, 3, 3)

ax[0].imshow(img, cmap=plt.cm.gray)
ax[0].set_title('before')
ax[0].axis('off')

ax[1].imshow(magnitude_spectrum, cmap=plt.cm.gray)
ax[1].set_title('ft')
ax[1].axis('off')


ax[2].imshow(os, cmap=plt.cm.gray)
ax[2].set_title('after')
ax[2].axis('off')

plt.show()

```
