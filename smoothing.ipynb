```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('images.jpg')

blur = cv2.blur(img,(5,5))

plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```

# analysis:
we are using avg filter by blur() function and the size of this mask is 5 the the output was image without noises , the details of edges is hiden  


# median filter
we are using median filter by function called medianblur then the output was image without noises and the details of edges is cleary

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img2 = cv2.imread('images.jpg')#read the image

median = cv2.medianBlur(img2,5)#median filter on the image with size mask 5

plt.subplot(121),plt.imshow(img2),plt.title('Original ')#drawing the image before and after
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(median ),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```

# Gaussian Blurring
In this, instead of box filter, gaussian kernel is used. It is done with the function, cv2.GaussianBlur(). We should specify the width and height of kernel which should be positive and odd. 

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img3 = cv2.imread('images.jpg')#read the image

gg = cv2.GaussianBlur(img3,(5,5),0)
#gaussian filter on the image with size mask 5

plt.subplot(121),plt.imshow(img3),plt.title('Original ')#drawing the image before and after
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(gg ),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```
